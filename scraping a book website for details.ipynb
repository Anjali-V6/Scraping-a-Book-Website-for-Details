```python
import requests
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
URL = 'https://books.toscrape.com/'
response = requests.get(URL)

page_contents = response.text
```

```python
doc = BeautifulSoup(page_contents,'html.parser')
```

```python
Book_title_tags = doc.find_all('h3')
titles = []
for tags in Book_title_tags:
    titles.append(tags.text)

titles    
```

```python
Book_price_tags = doc.find_all('p', class_ = 'price_color')
price = []
for tags in Book_price_tags:
    price.append(tags.text.replace('Â',''))


price
```

```python
Book_stock_tags = doc.find_all('p', class_ = 'instock availability')
stock = []
for tags in Book_stock_tags:
    stock.append(tags.text.strip())
    
stock
```

```python
urls = []
for article in Book_title_tags:
    for link in article.find_all('a', href = True):
        url = link['href']
        links = 'https://books.toscrape.com/' + url
        if links not in urls:
            urls.append(links)

urls
```

```python
book_details = {'TITLE':titles,
                'PRICE':price,
                'STOCK AVAILABILTY':stock,
                'URL':urls}
book_details
```

```python
df = pd.DataFrame(book_details)
```
